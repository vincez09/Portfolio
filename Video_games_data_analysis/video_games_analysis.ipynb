It can be seen that there are null values and object data type for columns 'Times Listed', 'Number of Reviews', 'Plays', 'Playing', 'Backlogs', 'Wishlist' which should be a int or float.

I will clean the data by removing the null values and converting the said columns to int to use it for analysis.

## Data Cleaning

**Drop Unnamed Column**

```python
data = data.drop(columns = 'Unnamed: 0')
```

**Drop all duplicates**

```python
data[data.duplicated() == True]
```

```python
data.drop_duplicates(inplace = True)
```

**Convert row values that contains 'K' to int on Columns Times Listed, Number of Reviews, Plays, Playing, Backlogs, and Wishlist**

```python
def k_remover_and_convert(x):
    if 'K' in str(x):
        return float(x.replace('K', '')) * 1000
    return x

columns_to_process = ['Times Listed', 'Number of Reviews', 'Plays', 'Playing', 'Backlogs', 'Wishlist']

data.loc[:, columns_to_process] = data.loc[:, columns_to_process].applymap(k_remover_and_convert)
data[columns_to_process] = data[columns_to_process].astype(int)
```

```python
data.head(3)
```

**Convert Release Date column data type into string.**

```python
data['Release Date'] = data['Release Date'].astype(str)
```

**Drop rows having null values**

I will not impute missing values using mean or other methods because Rating varies a lot depending the quality of game. For team and summary, i will just impute it as 'blank'

```python
data.isna().sum()
```

```python
data.loc[data['Team'].isna(), 'Team'] = "['none']"
```

```python
data.loc[data['Summary'].isna(), 'Summary'] = "['none']"
```

```python
data.info()
```

```python
data.dropna(axis = 0, inplace=True)
```

## Game Popularity Analysis

**Top 20 Video Games by Number of Reviews**

```python
sorted_data = data[['Title', 'Number of Reviews']].sort_values(by='Number of Reviews', ascending=False)

# Plot the data
plt.figure(figsize=(10, 6))
bar_width = 0.6
plt.barh(sorted_data['Title'].head(20), sorted_data['Number of Reviews'].head(20), color='royalblue', height = bar_width)
plt.xlabel('Number of Reviews')
plt.ylabel('Title')
plt.title('Top 20 Video Games by Number of Reviews')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest number of reviews at the top
plt.show()
```

**Top 20 Video Games with Number of Reviews > 2000**

Some video games have high rating but with low number of reviews making it unreliable. So I will only include video games with number of reviews > 2000.

```python
top_20_games = data[data['Number of Reviews'] > 2000].sort_values('Rating', ascending=False).head(20)

# Plot the data
plt.figure(figsize=(10, 8))
bar_width = 0.6
bars = plt.barh(top_20_games['Title'], top_20_games['Rating'], color='royalblue', height = bar_width)

# Add data labels (number of reviews)
for bar, reviews in zip(bars, top_20_games['Number of Reviews']):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, reviews, ha='left', va='center')

plt.xlabel('Rating')
plt.ylabel('Title')
plt.title('Top 20 Video Games by Rating with Number of Reviews > 2000')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest rating at the top
plt.show()
```

**Top 20 Video Games by Number of Plays**

```python
game_plays = data[['Title', 'Plays']].sort_values(by='Plays', ascending=False).head(20)

# Plot the data
plt.figure(figsize=(10, 6))
bar_width = 0.6
plt.barh(game_plays['Title'], game_plays['Plays'], color='royalblue', height = bar_width)
plt.xlabel('Number of Plays')
plt.ylabel('Title')
plt.title('Top 20 Video Games by Number of Plays')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest number of reviews at the top
plt.show()
```

## Genre Analysis

**Showing 10 highest number of video games made based on Genre**

```python
genre_counts = data['Genres'].value_counts().head(10)

# Plot top 10 genres
plt.figure(figsize=(10, 6))
bar_width = 0.6
genre_counts.plot(kind='bar', color='royalblue', width = bar_width)
plt.title('Top 10 Video Games by Genre')
plt.xlabel('Genre')
plt.ylabel('Number of Games')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Top Highest Number of Plays based on Genres**

```python
exploded_data = data.explode('Genres')

# Group by genre and calculate the total number of plays for each genre
genre_plays = exploded_data.groupby('Genres')['Plays'].sum().sort_values(ascending=False).head(10)

# Plot the number of plays based on top 10 genres
plt.figure(figsize=(10, 6))
genre_plays.plot(kind='bar', color='royalblue')
plt.title('Number of Plays Based on Top 10 Genres')
plt.xlabel('Genre')
plt.ylabel('Total Number of Plays')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Correlation Analysis

**Correlation between the Rating and Number of Plays**

```python
top_20_highest_rated = data[['Title', 'Rating', 'Plays']].nlargest(20, 'Rating')

# Plot the data
plt.figure(figsize=(10, 8))
plt.scatter(data['Plays'], data['Rating'], color='royalblue')

# Add data labels for plays > 20000
for index, row in data[data['Plays'] > 20000].iterrows():
    plt.text(row['Plays'], row['Rating'], row['Title'], ha='left', va='center')

plt.xlabel('Plays')
plt.ylabel('Rating')
plt.title('Relationship between Number of Plays and Rating')
plt.grid(True)
plt.show()
```

In the scatterplot, it is evident that most video games with over 20,000 plays have high ratings. However, 'Among Us' stands out as an exception due to its low rating.

**Distribution of Game Ratings**

```python
rating_counts = data['Rating'].value_counts().sort_index()

# Plot the rating distribution as a line chart
plt.figure(figsize=(10, 6))
plt.plot(rating_counts.index, rating_counts.values, color='royalblue', linewidth=2)

# Add labels and title
plt.xlabel('Rating', fontsize=14, fontweight='bold')
plt.ylabel('Count', fontsize=14, fontweight='bold')
plt.title('Rating Distribution', fontsize=16, fontweight='bold')

# Customize x and y ticks
plt.xticks(range(0, 6), fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.tight_layout()
plt.grid(False)
plt.show()
```
